```python
%%writefile main.py
from fastapi import FastAPI
from pydantic import BaseModel

url = "https://github.com/DataTalksClub/mlops-zoomcamp/blob/main/cohorts/2023/04-deployment/web-service-model/pipeline_v1.bin?raw=true"
with urllib.request.urlopen(url) as f:
    model = pickle.load(f)

class ClientData(BaseModel):
    lead_source: str
    number_of_courses_viewed: int
    annual_income: float

app = FastAPI()

@app.get("/")
def home():
    return {"message": "Lead conversion prediction API is running"}

@app.post("/predict")
def predict(client: ClientData):
    data = client.dict()
    X = {
        "lead_source": [data["lead_source"]],
        "number_of_courses_viewed": [data["number_of_courses_viewed"]],
        "annual_income": [data["annual_income"]],
    }
    pred = model.predict_proba(X)[0, 1]
    return {"conversion_probability": round(float(pred), 3)}
```
